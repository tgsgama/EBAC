Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Importando as bibliotecas
import pandas as pd
import requests
import json
import numpy as np
````

````python
#1) Seu código aqui
def get_dados_api(uf: str = 'MG', tipo_interessado: str = 'Estado'):
    ti = tipo_interessado.casefold()
    uf = uf.upper()
    url_municipio = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Munic%C3%ADpio'
    url_estado = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Estado'
    
    if ti == 'estado':
        read = requests.get(url_estado)
        if read.status_code == 200:
            print(f'Requisição realizada com sucesso. Gerando o Dataframe...')
            data_dict = json.loads(read.text)
            df = pd.DataFrame(data_dict['items'])
            return df
        else:
            print(f'Não foi possivel realizar a requisição.')
            
    elif ti == 'municipio':
        read = requests.get(url_municipio)
        if read.status_code == 200:
            print(f'Requisição realizada com sucesso. Gerando o Dataframe...')
            data_dict = json.loads(read.text)
            df = pd.DataFrame(data_dict['items'])
            return df
    else:
        print('O (tipo_interessado) deve ser Estado ou Municipio')
        df = pd.DataFrame({'id_pleito' : [np.nan], 'tipo_interessado': [np.nan], 'interessado': [np.nan], 'cod_ibge': [np.nan],
                             'uf': [np.nan], 'num_pvl': [np.nan], 'status': [np.nan], 'num_processo': [np.nan], 'data_protocolo': [np.nan],
                            'tipo_operacao': [np.nan], 'finalidade': [np.nan], 'tipo_credor': [np.nan], 'credor': [np.nan], 'moeda': [np.nan],
                            'valor': [np.nan], 'pvl_assoc_divida': [np.nan], 'pvl_contradado_credor': [np.nan], 'data_status': [np.nan]})
        return df
        
````

````python
get_df = get_dados_api('MG', 'municipio')
get_df.head(n=25)
````

````python
# 2) Seu código aqui
filtro_df = get_df[get_df['status'] == 'Arquivado por decurso de prazo']
filtro_df['status'].value_counts()
````
